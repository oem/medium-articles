# Exploring data in julia

## Previously

Last time, we build a [webcrawler]() that would crawl the [hamburg.de](https://www.hamburg.de/corona-zahlen/) website for the current covid-19 numbers in hamburg.

I went ahead and wrote a small package that provides those numbers for all recorded days.

Let's see what we can squeeze out of that data!

```julia
using Hamburg
```

```julia
using Dates
using DataFrames

using Gadfly
Gadfly.set_default_plot_size(700px, 300px)
```

The first dataset shows the infections per borough, aggregated for 14 days. The aggregation happens to protect the privacy of the infected people.

We did already plot the data when we build the webcrawler, but let's quickly do it again, this time for all the rows.

```julia
boroughs = dataset("covid-19", "boroughs")
select!(boroughs, Not(:recordedat))
```

```julia
plot(x=names(boroughs), 
    y=convert(Matrix, boroughs)[1, :],
    Geom.bar,
    Guide.xlabel("Boroughs"),
    Guide.ylabel("New Infections"),
    Guide.title("New infections in the last 14 days"),
    Theme(bar_spacing=5mm))
```

## New stuff!

```julia
df = dataset("covid-19", "infected")
df[1:10, :]
```

```julia
plot(df, y=:new,x=:recordedat, Geom.bar, 
    Coord.cartesian(xmin=Date(2020, 3,1)), 
    Guide.title("Confirmed new COVID-19 cases in Hamburg, Germany"), 
    Guide.xlabel("date"), 
    Guide.ylabel("new cases"))
```
